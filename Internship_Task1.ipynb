```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.to_csv("iris.csv", index=False)
df
```

```python
print(f"Top Five rows are:\n",df.head())
print(f"Total rows and columns are:",df.shape)
print(f"Columns Names are\n",df.columns)
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df)
plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.show()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['petal_length'], df['petal_width'])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(6,4))
bins = [4.0,4.5,5.0,5.5,6.0,6.5,7.0,7.5,8.0]  # 4â€“5, 5â€“6, 6â€“7, 7â€“8
sns.histplot(df['sepal_length'],bins=bins, kde=True)
plt.title("Histogram: Sepal Length Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['petal_length'])
plt.title("Box Plot: Petal Length")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df.iloc[:, :4])
plt.title("Box Plot of All Numeric Features")
plt.show()
```
